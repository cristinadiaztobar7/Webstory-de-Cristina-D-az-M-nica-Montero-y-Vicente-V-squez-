```python
# =====================================================
# LIMPIAR TODOS LOS CSV EN /Datos/
# Conservar solo mujeres chilenas
# (usa coincidencias como "Damas", "Femenino", "(CHI)", "Chile")
# =====================================================

import pandas as pd
import glob
import os

# --- 1. Ruta donde están los archivos ---
carpeta = "/content/drive/MyDrive/Datos/"
archivos = glob.glob(os.path.join(carpeta, "pan-am-*-results.csv"))

print(f"Se encontraron {len(archivos)} archivos CSV.")

# --- 2. Procesar y limpiar cada archivo ---
todos = []

for archivo in archivos:
    print(f"Procesando: {os.path.basename(archivo)}")
    df = pd.read_csv(archivo)

    # Normalizar nombres de columnas
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
    )

    # Detectar columna que contiene "damas" o "femenino" or "mujer"
    col_genero = None
    for c in df.columns:
        if df[c].astype(str).str.contains("damas|femenino|mujer", case=False, na=False).any():
            col_genero = c
            break

    # Create gender filter - initialize as all False
    filtro_mujeres = pd.Series(False, index=df.index)
    if col_genero and col_genero in df.columns:
        filtro_mujeres = df[col_genero].astype(str).str.contains("damas|femenino|mujer|women", case=False, na=False)
    else:
        # If no specific gender column found, try searching in all string columns for gender keywords
        for col in df.select_dtypes(include='object').columns:
             filtro_mujeres |= df[col].astype(str).str.contains("damas|femenino|mujer|women", case=False, na=False)


    # Buscar columna con nombres para detectar "CHI"
    col_nombre = None
    for c in df.columns:
        if "nombre" in c:
            col_nombre = c
            break

    # Buscar columna país (de competencia o atleta)
    col_pais = None
    for c in df.columns:
        if "pais" in c or "país" in c or "country" in c: # Include 'country' in search
            col_pais = c
            break

    # Create chilean filter - initialize as all False
    filtro_chilenas = pd.Series(False, index=df.index)

    # Apply filters if columns are found
    if col_nombre and col_nombre in df.columns:
        filtro_chilenas = df[col_nombre].astype(str).str.contains("CHI", case=False, na=False)

    if col_pais and col_pais in df.columns:
         # Use element-wise OR with the existing filtro_chilenas Series
        filtro_chilenas = filtro_chilenas | df[col_pais].astype(str).str.contains("Chile|CHI", case=False, na=False)


    # Apply filters
    # Ensure both filters are boolean Series before applying
    if isinstance(filtro_mujeres, pd.Series) and isinstance(filtro_chilenas, pd.Series) and len(filtro_mujeres) == len(df) and len(filtro_chilenas) == len(df):
        filtrado = df[filtro_mujeres & filtro_chilenas].copy()
        if 'fuente' not in filtrado.columns:
             filtrado['fuente'] = os.path.basename(archivo)  # registrar de qué año viene
        else:
             # Append to existing 'fuente' if already present
             filtrado['fuente'] = filtrado['fuente'].astype(str) + ";" + os.path.basename(archivo)

        # Agregar a la lista
        if not filtrado.empty:
           todos.append(filtrado)
        else:
            print(f"No se encontraron mujeres chilenas en: {os.path.basename(archivo)}")
    else:
        print(f"Skipping {os.path.basename(archivo)} due to incompatible filter types or lengths.")


# --- 3. Unir todos los resultados en un solo DataFrame ---
if todos:
    chilenas = pd.concat(todos, ignore_index=True)

    # --- 4. Guardar el conjunto limpio ---
    salida = os.path.join(carpeta, "mujeres_chilenas_1951_2023.csv")
    chilenas.to_csv(salida, index=False, encoding='utf-8-sig')

    print(f"\n✅ Limpieza completada.")
    print(f"Archivo final guardado en: {salida}")
    print(f"Total de filas: {len(chilenas)}")
    display(chilenas.head())
else:
    print("\n❌ No se encontraron mujeres chilenas en ninguno de los archivos procesados.")
    chilenas = pd.DataFrame() # Create an empty DataFrame if no data is found
```
