```python
import pandas as pd
import glob
import os

# Ruta de la carpeta donde están los CSV
path = "/content/Webstory-de-Cristina-D-az-M-nica-Montero-y-Vicente-V-squez-/Entrega_03/Montero_Monica_vis_01/BASE DATOS HOMBRES LIMPIA POR AÑO"

# Buscar todos los archivos CSV dentro de esa carpeta
all_files = glob.glob(os.path.join(path, "*.csv"))

# Leer y concatenar, manteniendo todas las columnas posibles
df_list = []
for f in all_files:
    try:
        temp = pd.read_csv(f)
        temp["año"] = os.path.basename(f).split(".")[0]  # agrega columna con el año del archivo
        df_list.append(temp)
    except Exception as e:
        print(f"⚠️ Error al leer {f}: {e}")

df = pd.concat(df_list, ignore_index=True, sort=False)

# Mostrar las columnas detectadas y primeras filas
print("Columnas encontradas:", df.columns.tolist())
df.head()
```

```python
# Convertir a minúsculas y eliminar espacios en nombres de columnas
df.columns = df.columns.str.strip().str.lower()

# Quitar filas completamente vacías
df = df.dropna(how='all')

# Limpiar la columna 'mark' (quitar comas y letras)
df['mark'] = df['mark'].astype(str).str.replace(',', '.').str.replace('[^0-9.:]', '', regex=True)

# Filtrar solo atletas de Chile (si aplica)
df_chile = df[df['country'].str.contains('Chile', case=False, na=False)]

# Mostrar resumen
print("Total registros:", len(df_chile))
df_chile.head(10)
```

```python
import altair as alt
import pandas as pd

# Aseguramos que 'año' sea numérico
df_chile['año'] = pd.to_numeric(df_chile['año'], errors='coerce')

# Filtramos filas con marcas válidas (quitamos vacíos o texto extraño)
df_chile = df_chile[df_chile['mark'].str.contains(r'\d', na=False)]

# Seleccionamos algunos eventos representativos (puedes cambiarlos)
eventos_interes = [
    "100 METRES", "200 METRES", "400 METRES",
    "1,500 METRES", "5,000 METRES", "10,000 METRES",
    "HIGH JUMP", "LONG JUMP", "HAMMER THROW", "SHOT PUT"
]

df_viz = df_chile[df_chile['event'].isin(eventos_interes)]

# Crear gráfico con Altair
chart = (
    alt.Chart(df_viz)
    .mark_circle(size=70, opacity=0.7)
    .encode(
        x=alt.X('año:O', title='Año'),
        y=alt.Y('mark:Q', title='Marca (segundos o metros)', scale=alt.Scale(zero=False)),
        color=alt.Color('event:N', title='Evento'),
        tooltip=['name', 'event', 'mark', 'año', 'location']
    )
    .properties(
        title='Evolución de las marcas de atletas chilenos en los Juegos Panamericanos (1951–2023)',
        width=800,
        height=400
    )
    .interactive()
)

chart.save('vis_01.html')  # guarda versión interactiva
chart.save('vis_01.png')   # guarda versión en imagen
chart
```
